```python
ListOfAll=[]
for emoj in Emojis:
    ListOfAll.append(emoj)
    for H in Hair:
        ListOfAll.append(emoj+"\U0000200D"+H+TerminationChar)       
    for col in Tone:
        ListOfAll.append(emoj+col+TerminationChar)
        for H in Hair:
            ListOfAll.append(emoj+col+"\U0000200D"+H+TerminationChar)
                
    
```

```python
for a in ListOfAll: 
    print(a,list(grapheme.graphemes(a)))
```
